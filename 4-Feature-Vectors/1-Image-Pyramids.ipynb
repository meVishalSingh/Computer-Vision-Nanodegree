## Image pyramids

### Importing the Resources

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

### Displaying the Image

```python
# Read in the image
image = cv2.imread('images/rainbow_flag.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image);
```

### Downsampling the Image

```python
level_1 = cv2.pyrDown(image)
level_2 = cv2.pyrDown(level_1)
level_3 = cv2.pyrDown(level_2)

# Display the images
f, (ax1,ax2,ax3,ax4) = plt.subplots(1, 4, figsize=(20,10), gridspec_kw={'width_ratios':[1, 0.5, 0.25, 0.125]})

ax1.imshow(image);
ax1.axis('off');

ax2.imshow(level_1);
ax2.axis('off');

ax3.imshow(level_2);
ax3.axis('off');

ax4.imshow(level_3);
ax4.axis('off');
```
