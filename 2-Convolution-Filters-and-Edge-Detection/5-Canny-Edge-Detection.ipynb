# Canny Edge Detection

These are great for eliminating weak edges and noise and isolating edges that are a part of an object boundary.

### Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

### Displaying the Image

```python
# Read in the image
image = cv2.imread('images/brain_mr.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image);
```

```python
# Convert the image to grayscale for processing
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray');
```

### Implementing Canny Edge Detection

```python
# Defining the Thresholds
wide = cv2.Canny(gray, 30, 100)
tight = cv2.Canny(gray, 200, 240)
 
    
# Displaying the images
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
ax1.set_title('Wide')
ax1.imshow(wide, cmap='gray');
ax2.set_title('Tight')
ax2.imshow(tight, cmap='gray');
```

### Edge Detection on a Different Image

```python
# Read in the image
image = cv2.imread('images/sunflower.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image);
```

```python
# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
# Trying different Thresholds
small = cv2.Canny(gray, 0, 50)
wide = cv2.Canny(gray, 30, 100)
tight = cv2.Canny(gray, 200, 240)
subtle = cv2.Canny(gray, 120, 240)

f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(20,5))
ax1.set_title('Small')
ax1.imshow(small, cmap='gray');
ax2.set_title('Wide')
ax2.imshow(wide, cmap='gray');
ax3.set_title('Tight')
ax3.imshow(tight, cmap='gray');
ax4.set_title('Subtle')
ax4.imshow(subtle, cmap='gray');
```

As it can be seen above for a very small threshold a lot of noise will be picked up. A wide spectrum will shows a lot of detected edges whereas, a tight thereshold will show fewer edges as the cutoff for weak edges is higher. 

Usually, a ratio of **1:2** or **1:3** is adequate for the threshold range.
